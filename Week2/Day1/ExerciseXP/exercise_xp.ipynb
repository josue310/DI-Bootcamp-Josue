# 🌟 Exercise 6: Identifying Data Types
1. A company’s financial reports stored in an Excel file.
Réponse : Données structurées (Structured data)

Raisonnement : Un fichier Excel est organisé en lignes, en colonnes et en cellules avec des types de données précis (chiffres, dates, texte court). C'est le propre d'une structure tabulaire.

2. Photographs uploaded to a social media platform.
Réponse : Données non structurées (Unstructured data)

Raisonnement : Une image est un ensemble de pixels. Elle ne contient pas de structure interne native (comme un tableau) qui dit explicitement ce qu'elle représente. Pour un ordinateur, c'est un fichier brut.

3. A collection of news articles on a website.
Réponse : Données non structurées (Unstructured data)

Raisonnement : Même si les articles sont écrits avec des paragraphes, le texte libre (le langage naturel) est considéré comme non structuré car sa longueur varie et il n'est pas confiné dans des cases de base de données standard.

4. Inventory data in a relational database.
Réponse : Données structurées (Structured data)

Raisonnement : Une base de données relationnelle (SQL) utilise des tables strictes avec des clés primaires, des clés étrangères et des types de colonnes bien définis. C'est l'exemple parfait de la donnée structurée.

5. Recorded interviews from a market research study.
Réponse : Données non structurées (Unstructured data)

Raisonnement : Les enregistrements audio contiennent de la voix, des intonations, des bruits de fond et des silences. Il n'y a aucune structure tabulaire là-dedans; c'est un flux de données continu et complexe.
