```python
# 🌟 Exercise 2: Descriptive Statistics
import scipy.stats as stats
import numpy as np
data = [12, 15, 13, 12, 18, 20, 22, 21]

# mean
mean_data = stats.tmean(data)
# median
medium_data = np.median(data)
print(f" variance : {medium_data}")
#variance
variance_data = stats.tvar(data)

#standard deviation
std_data = stats.tstd(data)
print(std_data)
```

```python
# 🌟 Exercise 3: Understanding Distributions

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

means, std = 50, 10
normal_dis = stats.norm(means, std)
# PDF =probability density function
print(f"PDF at x=0.5 : {normal_dis.pdf(0.5)}")

# je definis 100 points
x = np.linspace(-0.4,0.4,100)

# je trace les points
plt.plot(x, normal_dis.pdf(x))
plt.title("Distribution normale")
plt.show()

```

```python
# 🌟 Exercise 4: T-Test Application
import scipy.stats as stats

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_statistics, p_value = stats.ttest_ind(data1, data2)
print("T-statistic : ",t_statistics)
print("P-value : ", p_value)
```

```python
# 🌟 Exercise 5: Linear Regression Analysis

import numpy as np
from scipy import stats

x = np.array([50, 70, 80, 100, 120])
y = np.array([12, 24, 37, 43, 50])

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
print("The slope is : ",slope)
print("The intercept is : ",intercept)

# price of the house that is 90 square meters
value = intercept + 90 * slope
print("the price's prediction is : ",value)

# Interpret the meaning of the slope in the context of housing prices
print("La pente dans ce contexte correspond au coefficient de variation du prix d'une maison en fonction d'un metre carré")
```

```python
# Exercice 6

from scipy import stats

fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print("F_value : ",f_value)
print("P_value : ",p_value)

# Question 1 : Based on the P-value, do the fertilizers have significantly different effects on plant growth?
print(f"La P-valeur {p_value} est largement inférieure au seuil classique de 0,05. Par conséquent, oui, les fertilisants ont un effet significativement différent sur la croissance des plantes (l'un d'eux au moins est plus efficace que les autres")

# Explain what would happen if the P-value were greater than 0.05
print("On va dire que les groupes de plantes sont identiques donc les mesures sont sensiblement les mêmes malgré les types de fertilisants appliqués d'où les fertilisants ont un effet négligeable")
```

```python
# 🌟 Exercise 7: Probability Distributions (Optional)

from scipy import stats
import numpy as np

n, p = 10, 0.8
binomial_dist = stats.binom(n, p)
success_counts = np.arange(0, 11)
probabilities = binomial_dist.pmf(success_counts)

print("Nombre de succès | Probabilité")
print("-" * 30)
for success, prob in zip(success_counts, probabilities):
    print(f"{success:16d} | {prob:.4f} ({prob*100:.1f}%)")
```

```python
# 🌟 Exercise 8: Correlation Coefficients (Optional)

import pandas as pd
from scipy import stats

# 1. Création du DataFrame fourni
data = pd.DataFrame(
    {"age": [23, 25, 30, 35, 40], "income": [35000, 40000, 50000, 60000, 70000]}
)

# 2. Calcul du coefficient de Pearson
pearson_coef, pearson_p = stats.pearsonr(data["age"], data["income"])

# 3. Calcul du coefficient de Spearman
spearman_coef, spearman_p = stats.spearmanr(data["age"], data["income"])

# Affichage des résultats
print(f"Pearson correlation : {pearson_coef:.2f} (p-value: {pearson_p:.4f})")
print(f"Spearman correlation: {spearman_coef:.2f} (p-value: {spearman_p:.4f})")
```
