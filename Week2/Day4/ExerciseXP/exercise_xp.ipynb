```python
# 🌟 Exercise 2: Creating a Line Plot for Temperature Variation

import matplotlib.pyplot as plt

x = ["lundi", "mardi", "mercredi", "jeudi", "vendredi", "samedi", "dimanche"]

y = [72, 74, 76, 80, 82, 78, 75]

plt.title("temperature variation")
plt.xlabel('Day')
plt.ylabel('Temperature (°F)')
plt.plot(x, y)
plt.show()
```

```python
# 🌟 Exercise 4: Visualizing the Distribution of CGPA

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# charger le dataset

df = pd.read_csv('Student_Mental_health.csv')


sns.histplot(df['What is your CGPA?'], bins = 20, color = 'skyblue')
plt.title('Distribution of CGPA')
plt.show()
```

```python
# 🌟 Exercise 5: Comparing Anxiety Levels Across Different Genders

import seaborn as sns
import matplotlib.pyplot as plt

print(df.dtypes)
sns.barplot(x='Choose your gender', y='Do you have Anxiety?', data=df, palette='Blues', hue='Choose your gender')
plt.title("Anxiety Level Across Different Genders")
plt.xlabel("Gender")
plt.ylabel("Proportion of Anxiety")
plt.show()
```

```python
# 🌟 Exercise 6: Exploring the Relationship Between Age and Panic Attacks

import matplotlib.pyplot as plt
import seaborn as sns

# Conversion Yes/No en numérique
df['Panic_Attacks_Num'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Panic_Attacks_Num',
                s=100, alpha=0.6, color='crimson', edgecolor='black')

# Personnalisation
plt.title('Relation entre l\'âge et les crises de panique', fontsize=14)
plt.xlabel('Âge', fontsize=12)
plt.ylabel('Crises de panique (0 = Non, 1 = Oui)', fontsize=12)
plt.yticks([0, 1], ['Non', 'Oui'])
plt.grid(True, linestyle='--', alpha=0.4)

plt.show()
```
